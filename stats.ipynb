```python
import os, json 
from pandas import Series

with open("legend.json", "r") as f:
    legend = json.load(f)

with open("labels.json", "r") as f:
    labels = json.load(f)
```

```python
total_number_of_label_groups = len(legend)
total_number_of_labels = 0
reverse_labels = {}
for group in legend:
  total_number_of_labels += len(legend[group])
  reverse_labels |= {label: key for key, label in legend[group].items()}

print("Total number of label groups:", total_number_of_label_groups)
print("Total number of distinct labels:", total_number_of_labels)
```

```python
label_distribution = Series(labels).value_counts().reset_index().rename(columns={"index": "label"})
label_distribution["label_name"] = label_distribution["label"].map(reverse_labels)

label_distribution[:60]
```
